## 1. Why is there no "Training" Phase?

This project implements a **Zero-Shot Hybrid Pipeline** for document extraction. Instead of training a custom layout-aware model (like LayoutLM) from scratch—which requires thousands of manually annotated bounding boxes and significant GPU compute—we utilize a deterministic rule-based spatial engine combined with a powerful pre-trained Large Language Model (`gemini-2.5-flash`).

**Architecture Recap:**
1. **Pre-trained OCR:** We use PaddleOCR (pre-trained) to extract raw text tokens and bounding boxes.
2. **Spatial Engine:** A deterministic algorithmic layer that performs Y-axis row grouping and X-axis column clustering. This requires no ML training.
3. **Zero-Shot LLM Extraction:** The structured spatial data is passed into Google Gemini with a zero-shot prompt and Pydantic schema constraints. The LLM relies on its vast pre-trained knowledge to semantically map the fields without needing fine-tuning.

Because of this architectural choice, **no explicit gradient descent, backpropagation, or model weight updates (training) occur in this repository.**

Instead, this notebook demonstrates how we **evaluate** the pipeline's performance on synthetic GST data and the public SROIE dataset.
